```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is in a DataFrame called 'df'
df = pd.read_csv("Downloads//gender_submission.csv")

# Select the column for which you want to create a histogram
selected_column = df['Survived']

# Create a histogram
plt.hist(selected_column, bins=10)  # You can adjust the number of bins as needed

# Add labels and a title
plt.xlabel('X-axis Label')
plt.ylabel('Frequency')
plt.title('Histogram of Data Distribution')

# Display the histogram
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Select the column for which you want to create a bar chart
selected_column = df['Survived']

# Count the occurrences of each category
value_counts = selected_column.value_counts()

# Create a bar chart
plt.bar(value_counts.index, value_counts.values)

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Bar Chart of Data Distribution')

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the bar chart
plt.show()
```
